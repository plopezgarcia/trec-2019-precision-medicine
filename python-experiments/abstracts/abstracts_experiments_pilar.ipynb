```python
for boost_disease in numpy.arange(1, 5):
    
    for boost_gene in numpy.arange(1, 5):
        
        print("boost_disease:", boost_disease)
        print("boost_gene:", boost_gene) 
        
        run_params = {
            'run_id':'DEFAULT_RUN',
            'query_template':'test_pilar/baseline_all_fields.json',
            'output_size' : 1000,

            'disease_tie_breaker': 0 ,
            'disease_multi_match_type':'best_fields',
            'disease_boost': boost_disease,

            'gene1_tie_breaker': 0 ,
            'gene1_multi_match_type':'best_fields',
            'gene1_boost': int(boost_gene),

            'gene2_tie_breaker': 0 ,
            'gene2_multi_match_type':'best_fields',
            'gene2_boost': int(boost_gene),

            'gene3_tie_breaker': 0,
            'gene3_multi_match_type':'best_fields',
            'gene3_boost': int(boost_gene),

        }

        run, run_params = running.run(topics_all, 'ABSTRACTS', run_params)
        results, aggregated = evaluation.evaluate(qrels[(qrels.relev.isin([2]))], run)
        print(aggregated)
```
